```python
# https://diegoinacio.medium.com/metricas-de-distancia-e-dissimilaridade-94f9d8d962d4#2f5d
def minkowski_distance(x, y, p):
    distance = np.sum(np.abs(x - y) ** p) ** (1 / p)
    return distance
```

```python
# https://diegoinacio.medium.com/metricas-de-distancia-e-dissimilaridade-94f9d8d962d4#8cf5
def manhattan_distance(x, y):
    distance = np.sum(np.abs(x - y))
    return distance
```

```python
# https://diegoinacio.medium.com/metricas-de-distancia-e-dissimilaridade-94f9d8d962d4#a8b3
def euclidean_distance(x, y):
    distance = np.sqrt(np.sum((x - y) ** 2))
    return distance
```

```python
# exemplos de dados X e Y
# quando usar cada função de distância: https://cursos.alura.com.br/forum/topico-quando-devo-usar-cada-funcao-de-distancia-327038#1394205
# três números aleatórios
x = np.random.randint(0, 10, size=3)
y = np.random.randint(0, 10, size=3)

print("Distância de Minkowski:", minkowski_distance(x, y, 3))
print("Distância de Manhattan:", manhattan_distance(x, y))
print("Distância Euclidiana:", euclidean_distance(x, y))
```

```python
# https://medium.com/@yucavalcante/s%C3%A9ries-temporais-com-python-quando-e-como-utilizar-08e1a95d1943
# Gráficos de séries temporais com Plotly

# Dados fictícios
np.random.seed(42)
data = pd.date_range(start="2023-01-01", periods=104, freq='W')
vendas = np.cumsum(np.random.randn(104)*10 + 0.5)

# Criando o gráfico de séries temporais
df = pd.DataFrame(data={"Vendas": vendas}, index=data)
ax = df.plot(figsize=(10, 6), color='red', linestyle='-', linewidth=2)
ax.set_title("Vendas Semanais da Apple", fontsize=16)
ax.set_xlabel("Datas", fontsize=14)
ax.set_ylabel("Vendas", fontsize=14)
ax.grid(True, which='both', linestyle='--', linewidth=0.5)
ax.legend(["Vendas"])
plt.tight_layout()
plt.show()
```
